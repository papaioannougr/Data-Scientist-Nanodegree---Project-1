We will keep only the columns that seem manageable and have interesting info.
With the term manage-able information we mean those categorical variables that have a relatively low number of unique values.

```python
df_list=pd.DataFrame(
df_listingsr[['id','host_response_time','host_response_rate','host_acceptance_rate','host_is_superhost','reviews_per_month',
'host_neighbourhood','host_listings_count','host_total_listings_count','host_has_profile_pic','host_identity_verified',
'neighbourhood_group_cleansed','smart_location','latitude','longitude','is_location_exact','property_type','room_type',
'accommodates','bathrooms','bedrooms','beds','bed_type','amenities','square_feet','price','weekly_price','monthly_price',
'security_deposit','cleaning_fee','guests_included','extra_people','minimum_nights','maximum_nights','calendar_updated',
'has_availability','availability_30','availability_60','availability_90','availability_365','number_of_reviews','first_review',
'last_review','review_scores_rating','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin',
'review_scores_communication','review_scores_location','review_scores_value','requires_license','instant_bookable',
'cancellation_policy','require_guest_profile_picture','require_guest_phone_verification','calculated_host_listings_count']]
)

df_list.describe(include='all')
```

These columns:

host_is_superhost
host_has_profile_pic
host_identity_verified
is_location_exact
has_availability
requires_license
instant_bookable
require_guest_profile_picture
require_guest_phone_verification
host_acceptance_rate

are essetntially booleans. So, we will create new boolean columns adding the suffix '_bool' in their names
and then we will drop these columns, keeping only the respective generated boolean columns

```python
list=['host_is_superhost',
'host_has_profile_pic',
'host_identity_verified',
'is_location_exact',
'has_availability',
'requires_license',
'instant_bookable',
'require_guest_profile_picture',
'require_guest_phone_verification']

for col in list:
    df_list[col+'_bool']=np.where(df_list[col]=='t', True, False)
    df_list.drop([col],axis=1,inplace=True)

    
df_list['host_acceptance_rate'+'_bool']=np.where(df_list['host_acceptance_rate']=='100%', True, False)
df_list.drop(['host_acceptance_rate'],axis=1,inplace=True)

df_list.describe(include='all')
```

These columns bear numeric information about prices in dollars:

price

weekly_price

monthly_price

security_deposit

cleaning_fee

extra_people

We will create a loop where we are going to create new variable in place of the above ones.
New variables will be numerical, without the $ sign in their values.

```python
import locale
locale.setlocale( locale.LC_ALL, 'en_US.UTF-8' ) 

list=['price',
'weekly_price',
'monthly_price',
'security_deposit',
'cleaning_fee',
'extra_people'
]

for var in list:
    lst=[]
    df_list[var]=df_list[var].fillna('$-1.00')
    for index, row in df_list.iterrows():
        lst.append(locale.atoi(row[var][1:-3]))    
    df_list[var+'_dollar']=lst
    print(df_list[[var,var+'_dollar']].head(2))
    df_list[var+'_dollar']=np.where(df_list[var+'_dollar']==-1, np.nan, df_list[var+'_dollar'])
    df_list.drop([var],axis=1,inplace=True)
    print(df_list[[var+'_dollar']].head(2))
```

We will apply the same concept in order to tranform column:
'host_response_rate' into a numerical, deleting the percentage sign 


```python
lst=[]
df_list.host_response_rate=df_list.host_response_rate.fillna('-1%')
for index, row in df_list.iterrows():
    lst.append(int(row.host_response_rate [:-1]))
df_list['host_response_rate_perc']=lst
print(df_list[['host_response_rate','host_response_rate_perc']].head(5))
df_list['host_response_rate_perc']=np.where(df_list['host_response_rate_perc']==-1, np.nan, df_list['host_response_rate_perc'])
df_list.drop(['host_response_rate'],axis=1,inplace=True)

df_list[['host_response_rate_perc']].head()


df_list.describe(include='all')
```
